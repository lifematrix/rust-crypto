```python
import secrets, time

N = 1_000_000
size = 32

t0 = time.perf_counter()
for _ in range(N):
    secrets.token_bytes(size)
t1 = time.perf_counter()

bytes_total = N * size
print("seconds:", t1 - t0)
print("MB/s:", (bytes_total / (t1 - t0)) / (1024*1024))
print("us/call:", (t1 - t0) * 1e6 / N)
```

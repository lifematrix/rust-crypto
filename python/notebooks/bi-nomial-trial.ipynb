```python
import numpy as np
import matplotlib.pylab as plt
```

```python
def generate_bimonial_sample(p=0.4, n=1000, k=10000, seed=12345):
    np.random.seed(seed)
    samples = []
    for i in range(k):
        one_sample = np.random.choice([1,0], size=n, p=[p, 1-p])
        samples.append(np.mean(one_sample))
    return samples    
```

```python
%time samples = generate_bimonial_sample(p=0.4, n=1067, k=200000)
```

```python
np.std(samples), np.mean(samples)
```
